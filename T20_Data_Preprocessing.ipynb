```python
import pandas as pd
import json 
```

```python
with open('t20_wc_match_results.json') as f:
    data = json.load(f)
    
```

```python
df_match_results = pd.DataFrame(data[0]['matchSummary'])
df_match_results.head()
```

```python
df_match_results.shape
```

```python
df_match_results.rename({'scorecard': 'msc_id'}, axis = 1, inplace = True)
df_match_results.head()
```

```python
df_match_results.to_csv('t20_preprocessed_match_summary.csv', index = False)
```

```python
msc_ids_dict = {}

for index, row in df_match_results.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    msc_ids_dict[key1] = row['msc_id']
    msc_ids_dict[key2] = row['msc_id']
```

```python
with open('t20_wc_batting_summary.json') as f:
    data = json.load(f)
```

```python
all_records = []
for rec in data:
    batting_summaries = rec.get('battingSummary', [])
    all_records.extend(batting_summaries)
```

```python
df_batting_summary = pd.DataFrame(all_records)
df_batting_summary.head(9)
```

```python
df_batting_summary['out'] = df_batting_summary.dismissal.apply(lambda x: 1 if len(x)>0 else 0)
df_batting_summary.head(11)
```

```python
df_batting_summary['msc_id'] = df_batting_summary['match'].map(msc_ids_dict)
df_batting_summary.head()
```

```python
df_batting_summary.drop(columns=["dismissal"], inplace=True)
df_batting_summary.head(10)
```

```python
df_batting_summary['batsmanName'] = df_batting_summary['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting_summary['batsmanName'] = df_batting_summary['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting_summary.head()
```

```python
df_batting_summary.shape
```

```python
df_batting_summary.to_csv('t20_preprocessed_bating_summary.csv', index = False)
```
